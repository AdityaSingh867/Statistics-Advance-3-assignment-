Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

the process by which the numerical value of unknown population values are inferred from incomplete data, such as a sample.

A point estimate is a single value estimate of a parameter. For instance, a sample mean is a point estimate of a population mean. An interval estimate gives you a range of values where the parameter is expected to lie. A confidence interval is the most common type of interval estimate.

Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.

```python
def estimate_pop_mean(sample_mean, sample_std, sample_size):
    
    confidence_level = 0.95
    
    degrees_of_freedom = sample_size - 1
    
    t_value = abs(stats.t.ppf((1 - confidence_level) / 2, degrees_of_freedom))
    
    margin_of_error = t_value * (sample_std / math.sqrt(sample_size))
    
    lower_bound = sample_mean - margin_of_error
    
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, sample_mean, upper_bound
```

```python
import math
from scipy import stats

sample_mean = 24.5
sample_std = 5.1
sample_size = 100

lower_bound, pop_mean_estimate, upper_bound = estimate_pop_mean(sample_mean, sample_std, sample_size)
print("The estimated population mean is between", round(lower_bound, 2), "and", round(upper_bound, 2))
print("The point estimate of the population mean is", round(pop_mean_estimate, 2))
```

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

```python
import math
from scipy import stats

alpha = 0.05

x1 = [5, 9, 11, 7, 10, 8, 6, 12, 7, 10]
x2 = [7, 10, 8, 12, 11, 9, 6, 10, 7, 8]

n1 = len(x1)
n2 = len(x2)
mean1 = sum(x1) / n1
mean2 = sum(x2) / n2
var1 = sum([(x - mean1) ** 2 for x in x1]) / (n1 - 1)
var2 = sum([(x - mean2) ** 2 for x in x2]) / (n2 - 1)
s1 = math.sqrt(var1)
s2 = math.sqrt(var2)

t = (mean1 - mean2) / math.sqrt((var1 / n1) + (var2 / n2))

df = n1 + n2 - 2

p_value = 2 * stats.t.cdf(-abs(t), df)

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.

```python
import math
from scipy import stats


alpha = 0.05


x1 = [5, 9, 11, 7, 10, 8, 6, 12, 7, 10]
x2 = [7, 10, 8, 12, 11, 9, 6, 10, 7, 8]


n = len(x1)
mean1 = sum(x1) / n
mean2 = sum(x2) / n
var1 = sum([(x - mean1) ** 2 for x in x1]) / (n - 1)
var2 = sum([(x - mean2) ** 2 for x in x2]) / (n - 1)
s_p = math.sqrt(((n - 1) * var1 + (n - 1) * var2) / (2 * (n - 1)))
t = (mean1 - mean2) / (s_p * math.sqrt(2/n))


df = 2 * (n - 1)


p_value = 2 * stats.t.cdf(-abs(t), df)


if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
